## Parts 3b,3c


```python
import pandas as pd
data = pd.read_csv("/Users/aartijain/desktop/GE.csv") #Part 3b: loading the data into a dataframe
data.describe() #Part 3c: use describe() to get a rough overview of the data
```

## Part 3d

```python
#Printing the columns
print(data.columns)

#Printing the indexes
print(data.index)

#Plotting the date vs close price
axis = data.plot(x='Date',y='Close',kind = 'line')
axis.set_ylabel('Close price')

#Creating the pivot table
print(data.pivot_table)

#Plotting the date vs the volume 
data.plot(x='Date',y='Volume',kind = 'line')

#Using the groupby feature

data.groupby(pd.DatetimeIndex(pd.to_datetime(data['Date'])).year).mean()

```
